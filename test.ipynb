```python
import numpy as np
import matplotlib.pyplot as plt

# 数据
l = np.array([176.413, 182.13, 185.09, 190.66, 195.13, 
              199.46, 203.9, 208.3, 212.85, 217.09])
x = np.array([i+1 for i in range(len(l))])   # 横坐标

# 拟合直线
A = np.vstack([x, np.ones(len(x))]).T
k, b = np.linalg.lstsq(A, l, rcond=None)[0]

# 绘图
plt.plot(x, l, 'o', label='点') 
plt.plot(x, k*x + b, '-', label='拟合直线')
plt.legend()
plt.xlabel('l/mm')
plt.ylabel('频率/kHz')
plt.title('相位法测空气中声速')

# 显示斜率
print(f"斜率为：{k:.3f}")
```
